```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\lab\Dataset\yahoo_finance\AAPL_Montly_updates.csv")
data
```

```python
data.isna().sum()
```

```python
fig, ax = plt.subplots(figsize=(8,4), dpi=100)
data['Open'].plot(legend=True, ax=ax)
plt.title('Apple stock price (Open)')
plt.show()
```

```python
plt.figure(figsize = (8,4), dpi=100)
plt.plot(data['Open'], color = "y", label = "open")
plt.plot(data['Close'], color = "r", label = "Close")
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Apple stock price (Open v Close)")
plt.legend()
plt.show()
```
