```python
#removing missing values from data
data1 = data.dropna()
data1
```

```python
data1.columns
```

```python
#removing unwanted columns
data1 = data1.drop(['Unnamed: 0','Zip'], axis=1)
data1.head()
```

```python
#importing label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#converting categorical labels into numeric labels
data1['Address'] = le.fit_transform(data1['Address'])
data1.head()
```

```python
#creating independent variable
X = data1.drop(['Price','Lon','Lat'], axis=1)
X
```

```python
#creating dependent variable
y = data1['Price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#importing Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#fitting dataset
lr.fit(X_train,y_train)
```

```python
#predicting x test
y_pred = lr.predict(X_test)
y_pred
```

```python
#accuracy_score
np.round(lr.score(X_test,y_test)*100)
```
